```python
import json
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import cv2
```

```python
with open('/home/zhongzh/Experiment/U-Net_Cell_Detection/log.json','r',encoding='utf-8') as f:
    logs = json.load(f)
f.close()
log_enhance = logs[0]
loss = log_enhance['Loss']
log_n_en = logs[1]
loss_1 = log_n_en['Loss']
epoch = list(range(1,len(loss)+1))
plt.figure(figsize=(6,6))
l1 = plt.plot(epoch,loss,color = 'blue',label = 'Enhanced')
l2 = plt.plot(epoch,loss_1,color = 'green',label = 'No Enhanced')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
